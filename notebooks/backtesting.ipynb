```python
from src.portfolio.allocator import compute_allocation
from src.portfolio.backtester import run_backtest, compute_portfolio_metrics
from configs.feature_config import RAW_FEATURES, TARGET
import pandas as pd
import numpy as np

def _compute_listing_gain_pct(df: pd.DataFrame) -> pd.Series:
    return (
        (df[TARGET["listing_price"]] - df[TARGET["issue_price"]])
        / df[TARGET["issue_price"]]
        * 100
    )
t_min = 0.3933623644527928
alpha = 1.9877580169252183
params = load_config()['logistic_regression']
result = run_backtest('logistic_regression', t_min=t_min, alpha=alpha, verbose=False, **params)
pipeline = result["final_pipeline"]
X_val = result["holdout"]
end_date_col = RAW_FEATURES["ipo_end_date"]

# 2. predict probabilities on test set
pos_idx = list(pipeline.classes_).index(1)
val_probs = pipeline.predict_proba(X_val)[:, pos_idx]
actual_returns = _compute_listing_gain_pct(X_val)
val_df = pd.DataFrame({
    "prob": val_probs,
    "actual_return": actual_returns.values,
    "date": X_val[end_date_col].values,
})
# 3. group by decision day (listing date)
day_returns, daily_allocations = [], []
for day, day_df in val_df.groupby("date", sort=False):
    probs = day_df["prob"].values
    returns = day_df["actual_return"].values
    # 4. compute allocation
    weights = compute_allocation(probs, t_min=t_min, alpha=alpha)
    # 5. portfolio return for this day
    portfolio_return_day = float(np.sum(weights * returns))
    day_returns.append(portfolio_return_day)
    daily_allocations.append(weights.sum())

# 6. compute all financial metrics
metrics = compute_portfolio_metrics(day_returns, daily_allocations)

print("=" * 60)
print("HOLDOUT TEST SET SUMMARY")
print("=" * 60)
print(f"  t_min:              {t_min}")
print(f"  alpha:              {alpha}")
print(f"  Num days:           {metrics['num_days']}")
print(f"  Mean daily return:  {metrics['mean_daily_return']:+.2f}%")
print(f"  Win rate:           {metrics['win_rate']:.2%}")
print(f"  Avg allocation:     {metrics['avg_allocation']:.2f}")
print(f"  % Days traded:      {metrics['pct_days_traded']:.2%}")
print(f"  Volatility:         {metrics['volatility']:.2f}")
print(f"  Sharpe-like:        {metrics['sharpe_like']:.2f}")
print("=" * 60)
```

<h2>Strategy Analysis Experiments</h2>

```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import TimeSeriesSplit

from configs.feature_config import RAW_FEATURES, TARGET
from src.models.trainer import train
from src.portfolio.backtester import run_backtest, _compute_listing_gain_pct
from src.utils.utils import load_config, load_raw_dataset

# ── build prediction dataframe (all CV folds pooled) ─────────────────────────

config = load_config()
params = config["logistic_regression"]
cv_cfg = config["cv"]
threshold = config["target"]["listing_gain_threshold_perc"]

df = load_raw_dataset()
listing_date_col = RAW_FEATURES["listing_date"]
end_date_col = RAW_FEATURES["ipo_end_date"]

df[listing_date_col] = pd.to_datetime(df[listing_date_col]).dt.tz_localize(None)
df[end_date_col] = pd.to_datetime(df[end_date_col], errors="coerce").dt.tz_localize(None)

known = df.dropna(subset=[end_date_col])
known = known[known[listing_date_col] >= known[end_date_col]]
median_gap = (known[listing_date_col] - known[end_date_col]).median()
mask = df[end_date_col].isna()
df.loc[mask, end_date_col] = df.loc[mask, listing_date_col] - median_gap
df = df.sort_values(end_date_col).reset_index(drop=True)

holdout_start = len(df) - math.ceil(len(df) * cv_cfg["holdout_fraction"])
X_cv = df.iloc[:holdout_start]

splitter = TimeSeriesSplit(n_splits=cv_cfg["n_splits"], gap=cv_cfg["gap"])

all_preds = []
for tr_idx, val_idx in splitter.split(X_cv):
    X_train = X_cv.iloc[tr_idx]
    X_val = X_cv.iloc[val_idx]

    pipeline = train(X_train, "logistic_regression", listing_gain_threshold_perc=threshold, **params)
    pos_idx = list(pipeline.classes_).index(1)
    probs = pipeline.predict_proba(X_val)[:, pos_idx]
    returns = _compute_listing_gain_pct(X_val)
    target = (returns > threshold).astype(int)

    fold_df = pd.DataFrame({
        "p": probs,
        "actual_return": returns.values,
        "target": target.values,
        "date": X_val[end_date_col].values,
    })
    all_preds.append(fold_df)

pred_df = pd.concat(all_preds, ignore_index=True)
print(f"Prediction dataframe: {len(pred_df)} rows from {cv_cfg['n_splits']} folds")

# ══════════════════════════════════════════════════════════════════════════════
# 1. Bucket Analysis
# ══════════════════════════════════════════════════════════════════════════════

def analyze_probability_buckets(df, prob_col="p", return_col="actual_return", target_col="target"):
    bins = [0, 0.3, 0.4, 0.5, 0.6, 0.7, 1.0]
    df = df.copy()
    df["prob_bin"] = pd.cut(df[prob_col], bins=bins)

    summary = df.groupby("prob_bin", observed=False).agg(
        count=(prob_col, "count"),
        avg_return=(return_col, "mean"),
        win_rate=(target_col, "mean"),
    ).reset_index()

    print("\n=== Probability Bucket Analysis ===")
    print(summary.to_string(index=False))
    return summary

bucket_summary = analyze_probability_buckets(pred_df)

# ══════════════════════════════════════════════════════════════════════════════
# 2. Expected Return vs Probability (monotonicity check)
# ══════════════════════════════════════════════════════════════════════════════

def expected_return_curve(df, prob_col="p", return_col="actual_return"):
    bins = np.linspace(0, 1, 11)
    df = df.copy()
    df["bin"] = pd.cut(df[prob_col], bins=bins)

    curve = df.groupby("bin", observed=False)[return_col].mean().reset_index()
    print("\n=== Expected Return vs Probability ===")
    print(curve.to_string(index=False))

    fig, ax = plt.subplots(figsize=(8, 4))
    ax.bar(range(len(curve)), curve[return_col], tick_label=[str(b) for b in curve["bin"]])
    ax.set_xlabel("Probability Bin")
    ax.set_ylabel("Mean Return %")
    ax.set_title("Expected Return by Probability Bucket")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()

    return curve

return_curve = expected_return_curve(pred_df)

# ══════════════════════════════════════════════════════════════════════════════
# 3. Temporal Stability Check
# ══════════════════════════════════════════════════════════════════════════════

def temporal_stability_analysis(df, prob_col="p", return_col="actual_return", date_col="date"):
    df = df.copy().sort_values(date_col)
    midpoint = len(df) // 2
    early = df.iloc[:midpoint]
    late = df.iloc[midpoint:]

    print("\n=== Temporal Stability ===")
    for name, sub in [("Early Period", early), ("Late Period", late)]:
        print(f"\n  {name}:")
        print(f"    Count:       {len(sub)}")
        print(f"    Mean Return: {sub[return_col].mean():.4f}%")
        print(f"    Mean Prob:   {sub[prob_col].mean():.4f}")

temporal_stability_analysis(pred_df)

# ══════════════════════════════════════════════════════════════════════════════
# 4. Threshold Sensitivity (t_min comparison)
# ══════════════════════════════════════════════════════════════════════════════

def compare_thresholds(thresholds, alpha, model_name, params):
    results = []
    for t in thresholds:
        res = run_backtest(model_name, t_min=t, alpha=alpha, verbose=False, **params)
        results.append({
            "t_min": t,
            "mean_daily_return": res["mean_daily_return"],
        })

    result_df = pd.DataFrame(results)
    print("\n=== Threshold Comparison ===")
    print(result_df.sort_values("mean_daily_return", ascending=False).to_string(index=False))

    fig, ax = plt.subplots(figsize=(8, 4))
    ax.plot(result_df["t_min"], result_df["mean_daily_return"], marker="o")
    ax.set_xlabel("t_min")
    ax.set_ylabel("Mean Daily Return %")
    ax.set_title("Strategy Return vs Probability Threshold")
    ax.axhline(0, color="r", ls="--", lw=1)
    plt.tight_layout()
    plt.show()

    return result_df

threshold_df = compare_thresholds(
    thresholds=[0.3, 0.4, 0.5, 0.6, 0.7, 0.8],
    alpha=2.0,
    model_name="logistic_regression",
    params=params,
)
```
